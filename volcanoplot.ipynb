```python
from bioinfokit import analys, visuz
import pandas as pd
# load dataset as pandas dataframe
df = pd.read_csv('volcano_data.csv')
df = df[pd.to_numeric(df['p-value paired with no imputation'], errors='coerce').notnull()]
df = df.astype({'p-value paired with no imputation': float, "Fold change": float})

visuz.GeneExpression.volcano(df=df, lfc="Fold change", pv="p-value paired with no imputation", geneid="Gene Symbol",
    lfc_thr = (0.5, 0.5) ,gstyle=2, sign_line=True, xlm=(-3,3,0.5), ylm=(0,3.7,1.3), figtype='svg', axtickfontsize=10,
    axtickfontname='Verdana',genenames=({"Eml1":"Eml1", 'Kdm5d':'Kdm5d', 'Pacsin1':'Pacsin1', 'Hsf1':'Hsf1', 'Ckmt1':'Ckmt1', 'Adgrl3':'Adgrl3',
                                         'Pcgf1':'Pcgf1'}),show =True)
# plot will be saved in same directory (volcano.png)
# set parameter show=True, if you want view the image instead of saving
```
